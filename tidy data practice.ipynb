```r
not_cancelled<-flights%>% filter(!is.na(dep_delay),!is.na(arr_delay))
not_cancelled
```

```r
not_cancelled%>%count(dest)
#count(): counts the number of instances within each group of variables. 
```

```r
not_cancelled%>%group_by(dest)%>%summarise(n=length(dest))
#not_cancelled%>%group_by(dest)%>%summarise(n=n())  
```

```r
summarise(flights,delay=mean(dep_delay,na.rm=TRUE))
```

```r
# relationship between the distance and average delay for each location.
by_dest<-group_by(flights,dest) #dest . 
names(flights)
delay<-summarise(by_dest,average_dep=mean(dep_delay,na.rm=TRUE),average_arr=mean(arr_delay,na.rm=TRUE))
delay
```

```r
delay1<-summarise(by_dest,count=n(),average_dep=mean(dep_delay,na.rm=TRUE),average_arr=mean(arr_delay,na.rm=TRUE))
delay_filter<-filter(delay1,count>20,dest!="HNL")
```

```r
ggplot(data=delay_filter,mapping=aes(x=dist,y=delay)) + geom_point(aes(size=count),alpha=1/3)
```

```r
delays<-not_cancelled%>%group_by(tailnum)%>%summarise(delay=mean(arr_delay))
ggplot(delays,aes(x=delay))+geom_freqpoly(binwidth=10)
```

```r
delays<-not_cancelled%>%group_by(tailnum)%>%summarise(delay=mean(arr_delay,na.rm=TRUE),count=n())
ggplot(delays,aes(x=count,y=delay))+geom_point(alpha=1/3)
#variation decreases as the sample size increases. 
```
